```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=plt.imread("Mona_Lisa.jpg")

plt.imshow( img )
plt.gca().set_xlabel( 'x' ) # set the x-label of the current Axes (returned by the gca)
plt.gca().set_ylabel( 'y' ) # set the y-label of the current Axes
img.shape
```

```python
listImgShifted =[]
imgX, imgY =img.shape[0], img.shape[1]

for i in range(-10, 0):
    imgShifted = np.zeros(img.shape)
    imgShifted[0:imgX+i, :, :]=img[-1*i:imgX, :, :]

    listImgShifted.append(imgShifted)

for i in range(0,11):
    imgShifted = np.zeros(img.shape)
    imgShifted[i:imgX, :, :]=img[0:imgX-i, :, :]

    listImgShifted.append(imgShifted)

for imgS in listImgShifted:
    print(np.sum(np.where(imgS==0)))
    print(imgS.ndim)
```
